In this project, I created a small dataset of bacterial genes related to antibiotic resistance. I used Python to load the data, count the different resistance types, and display the results using a bar graph. This helped me understand how simple biological data can be analyzed and visualized using Python.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("gene_organism_resistance.csv - Sheet1.csv")
df
```

```python
df['Resistance_Type'].value_counts()
```

```python
df['Resistance_Type'].value_counts().plot(kind='bar')
plt.title("Distribution of Antibiotic Resistance Types")
plt.xlabel("Resistance Type")
plt.ylabel("Number of Genes")
plt.show()
```
