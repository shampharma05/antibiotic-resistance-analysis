```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
drug_names = [
  "Aspirin","Aspirin","Ibuprofen","Ibuprofen","Paracetamol",
  "Metformin","Metformin","Atorvastatin","Atorvastatin","Caffeine"
  ]
```

```python
target_proteins = [
  "COX1","COX2","COX1","COX2","COX3",
  "AMPK","Complex I","HMGCR","LDLR","Adenosine receptor"
  ]
```

```python
df = pd.DataFrame({
      "Drug_Name": drug_names,
          "Target_Protein": target_proteins
          })
print(df)
```

```python
counts = df["Drug_Name"].value_counts()
print(counts)
```

```python
counts.plot(kind="bar")

plt.xlabel("Drug Name")
plt.ylabel("Number of Targets")
plt.title("Drug vs Number of Target Proteins")

plt.show()
```
